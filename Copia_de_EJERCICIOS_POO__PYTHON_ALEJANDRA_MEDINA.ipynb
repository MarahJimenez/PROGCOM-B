```python
class Cafetera:
    def preparar_cafe(self):
        self.__moler_cafe()
        self.__hervir_agua()
        self.__mezclar()
        self.__servir()
        print("¡Café listo!")

    def __moler_cafe(self):
        print("Molendo el café...")

    def __hervir_agua(self):
        print("Hirviendo el agua...")

    def __mezclar(self):
        print("Mezclando café y agua...")

    def __servir(self):
        print("Sirviendo el café...")

cafetera = Cafetera()
cafetera.preparar_cafe()
```

```python
class Paciente:
    def __init__(self, id, oxigeno, vacunas, prueba_covid):
        self.id = id
        self.oxigeno = oxigeno
        self.vacunas = vacunas
        self.prueba_covid = prueba_covid
        self.diagnostico = None

    def evaluar(self):
        """Evalúa al paciente según los criterios establecidos."""
        if self.oxigeno < 90:
            self.diagnostico = "SALA COVID (oxígeno bajo)"
        elif self.vacunas < 2:
            self.diagnostico = "SALA COVID (pocas vacunas)"
        elif self.prueba_covid:
            self.diagnostico = "SALA COVID (prueba positiva)"
        else:
            self.diagnostico = "URGENCIAS"
        return self.diagnostico

    def mostrar_info(self):
        return f"Paciente {self.id} → {self.diagnostico}"


class SistemaTriaje:
    def __init__(self):
        self.pacientes = []
        self.sala_covid = []
        self.urgencias = []
        self.espera = []
        self.MAX_CAMILLAS_URGENCIAS = 10

    def agregar_paciente(self, paciente):
        self.pacientes.append(paciente)

    def procesar_triaje(self):
        for paciente in self.pacientes:
            resultado = paciente.evaluar()
            if "COVID" in resultado:
                self.sala_covid.append(paciente)
            else:
                if len(self.urgencias) < self.MAX_CAMILLAS_URGENCIAS:
                    self.urgencias.append(paciente)
                else:
                    self.espera.append(paciente)

    def mostrar_resultados(self):
        print("\n--- PACIENTES EN SALA COVID ---")
        for p in self.sala_covid:
            print(p.mostrar_info())

        print("\n--- PACIENTES EN URGENCIAS ---")
        for p in self.urgencias:
            print(p.mostrar_info())

        print("\n--- PACIENTES EN ESPERA ---")
        for p in self.espera:
            print(p.mostrar_info())


if __name__ == "__main__":
    import random

    sistema = SistemaTriaje()

    for i in range(1, 51):
        oxigeno = random.randint(85, 100)
        vacunas = random.randint(0, 3)
        prueba = random.choice([True, False])
        paciente = Paciente(i, oxigeno, vacunas, prueba)
        sistema.agregar_paciente(paciente)

    sistema.procesar_triaje()
    sistema.mostrar_resultados()
```

```python
import random

class SensorHumedad:
    def __init__(self, cantidad_tomas=20):
        self.cantidad_tomas = cantidad_tomas
        self.mediciones = []

    def registrar_tomas(self):
        self.mediciones = [random.randint(0, 10) for _ in range(self.cantidad_tomas)]

    def obtener_mediciones(self):
        return self.mediciones


class ActuadorGrifo:
    def __init__(self):
        self.grifo_abierto = False

    def abrir_grifo(self):
        if not self.grifo_abierto:
            print(" Grifo ABIERTO.")
            self.grifo_abierto = True

    def cerrar_grifo(self):
        if self.grifo_abierto:
            print(" Grifo CERRADO.")
            self.grifo_abierto = False

    def estado_grifo(self):
        return "abierto" if self.grifo_abierto else "cerrado"


class SistemaHumedad:
    def __init__(self):
        self.sensor = SensorHumedad()
        self.actuador = ActuadorGrifo()

    def ejecutar(self):
        self.sensor.registrar_tomas()
        mediciones = self.sensor.obtener_mediciones()

        print("=== INICIO DE MONITOREO ===\n")

        for i, humedad in enumerate(mediciones, 1):
            print(f"[Medición {i}] Humedad: {humedad}")
            if humedad < 3:
                print(" Alerta: Humedad BAJA.")
                self.actuador.abrir_grifo()
            elif 3 <= humedad <= 8:
                print("Humedad moderada.")
                if self.actuador.grifo_abierto:
                    self.actuador.cerrar_grifo()
            elif humedad > 8:
                print(" Alerta: Humedad ALTA.")
                if self.actuador.grifo_abierto:
                    self.actuador.cerrar_grifo()
                    print(" Activando sistema de absorción de agua residual.")

            print(f"→ Grifo: {self.actuador.estado_grifo()}\n")

        print("=== FIN DEL MONITOREO ===")

if __name__ == "__main__":
    sistema = SistemaHumedad()
    sistema.ejecutar()
```
